```python
!pip install folium      
import numpy as np
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
posi=pd.read_csv('location_county.csv')#读取文件中的信息
num=314 
lat=np.array(posi['Lat'][0:num])#选取纬度信息
lon=np.array(posi['Long_'][0:num])#选取经度信息
```

```python
data1=[[lat[i],lon[i]] for i in range(num)]#生成热力图需要的经纬度坐标
data1
```

```python
map_osm=folium.Map(location=[35,-90],zoom_start=5)
HeatMap(data1).add_to(map_osm)
file_path=r"Heatmap.html"
map_osm.save(file_path)
```
